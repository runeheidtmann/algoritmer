```python
def insertionSort(A):
    for j in range(1,len(A)):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i]>key:
            A[i+1] = A[i]
            i = i - 1
        A[i+1] = key
    
    return A

l = [3,2,1,4,3,2,14,13,12,11,10]
print(insertionSort(l))
        
```

```python
import numpy as np
def selectionSort(A):
    inList = A
    outList = []
    while len(inList) > 0:
        i = findSmallest(inList)
        outList.append(del(inList[i]))
    return outList
def findSmallest(L):
    smallest = np.inf
    for i in range(len(L)):
        if L[i] < smallest:
            smallest = i
    return smallest


l = [3,2,1,4,3,2,14,13,12,11,10]
print(insertionSort(l))
```
