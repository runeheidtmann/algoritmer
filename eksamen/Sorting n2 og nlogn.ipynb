```python
def insertionSort(A):
    for j in range(1,len(A)):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i]>key:
            A[i+1] = A[i]
            i = i - 1
        A[i+1] = key
    
    return A

l = [3,2,1,4,3,2,14,13,12,11,10]
print(insertionSort(l))
        
```
