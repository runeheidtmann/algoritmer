<h1>Hashing</h1>
Hashing er en måde at lave indsætning, sletninger og søgninger. Intet andet.

Her i undervisningen går vi ud fra, at vi hasher heltal op til k-grænse.
Hvis man vi hashe andre data-typer, så skal man først omdanne det til en hashcode.

Vi har altså et univers af mulige keys $U = \{0,1,2,3,...k-1\}$

Et dictionary gemmer en delmængde: $K \subseteq U$, fx $K = \{{2,3,4,8}\}$

```python
class Node:
    def __init__(self,key):
        self.next = None
        self.key = key
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

class Hashtable:
    def __init__(self,m):
        self.table = [None for x in range(m)]
    
    def insert(self,x):
        if self.table[x] == None:
            chain = LinkedList()
            chain.head = Node(x)
            chain.size += 1
            self.table[x] = chain
        else:
            chain = self.table[x]
            head = chain.head
            newHead = Node(x)
            newHead.next = head
            chain.head = newHead
            chain.size += 1
              

    def search(self,k):
        
        return self.table[k] is not None 
        
    def delete(self,x):
        if self.table[x] is not None:
            chain = self.table[x]
            chain.head = chain.head.next
            chain.size -= 1
            if chain.size < 1:
                self.table[x] = None
    
```

<h2>Ideen i praksis</h2>
Der laves en hash-funktion, som laver et objekt om til et heltal. Derefter indsættes det hashede objekt ind i det meget mindre array, som man kan searche i i konstant tid.

MEN der kan opstå konflikter hvis længden af array er mindre end antallet af objekter, der skal gemmes. Det kan man deale med på to måder.

<h3>Chaining</h3>
Hvis en adresse i en hashtable allerede er optaget, så kan man indsætte flere objekter på samme adresse blot ved at chaine objekterne samme i en datastruktur. Det kunne være en linkedList. Det giver giver så en søgetid på O(1)+O(længden af listen). Man kunne også indsætte et binært søgetræ i stedet for en linkedlist. Så er søgetiden logn.
    
<h3>Open Adressing</h3>
I stedet for chaining kunne man finde adresser der IKKE er optaget, hvis det oprindelige er optaget. Det er der forskellige strategier for.
