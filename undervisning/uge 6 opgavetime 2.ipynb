<h3>3: Implementer insertionsort ud fra pseudokoden i Cormen side 18</h3>

```python
def insertion_sort(alist, prints=False):
    i= 0
    for j in range(1,len(alist)):
    
        key = alist[j]
        i = j - 1
    
        while i >= 0 and alist[i] > key:
            alist[i+1] = alist[i]
            i -= 1
        
        alist[i+1] = key
        if prints:
            print(alist)
    
    return alist
```

<h3>2: Cormen 2.3-2</h3>

The key operation of the merge sort algorithm is the merging of two sorted
sequences in the ‚Äúcombine‚Äù step. We merge by calling an auxiliary procedure
$MERGE(A, p, q, r)$, where $A$ is an array and $p$, $q$, and $r$ are indices into the array
such that $p <= q < r$. The procedure assumes that the subarrays $A[p..q]$ and
$A[q+1..r]$ are in sorted order. It merges them to form a single sorted subarray
that replaces the current subarray $A[p..r]$.

```python
def merge(A,p,q,r):
    #indeces of arrays to run in forloop.
    n1 = q-p+1
    n2 = r-q
    
    #The first array is give A[0] -> A[q-p]
    L = [x for x in A[:n1]]
    #The second array is from A[q+1] -> A[r]
    R = [x for x in A[n1:r]]
    print(f"L{L}")
    print(f"R{R}")
    
    i = 0
    j = 0
    k = 0
    newlist = []
    while i < n1-1 and j < n2-1:
        if i <= len(L)-1 and L[i] <= R[j]:
            newlist.append(L[i])
            i += 1
            
        else:
            newlist.append(R[j])
            j += 1
        k += 1
 
    # Copy the remaining elements of L[], if there
    # are any
    while i < n1:
        newlist.append(L[i])
        i += 1
        k += 1
 
    # Copy the remaining elements of R[], if there
    # are any
    while j < n2-1:
        newlist.append(R[j])
        j += 1
        k += 1        

    print(newlist)
        
    
    
    
```

```python
from time import time
import matplotlib.pyplot as plt
import random
import numpy as np

def insertion_sort(alist, prints=False):
    i= 0
    for j in range(1,len(alist)):
    
        key = alist[j]
        i = j - 1
    
        while i >= 0 and alist[i] > key:
            alist[i+1] = alist[i]
            i -= 1
        
        alist[i+1] = key
        if prints:
            print(alist)
    return alist
```

```python
def runSortReturnTime(i):
    
    blist = [x for x in range(i,0,-1)]
    alist = [x for x in range(0,i,1)]
    
    countTime = 0
    for i in range(3):
        
        start = time()
        insertion_sort(alist)
        end = time()
        atime = end-start

        start = time()
        insertion_sort(blist)
        end = time()
        btime = end-start

        avgTime = (atime+btime)/2

        countTime += avgTime
    
    return countTime/3

```

```python
X = []
Y = []
for j in range(100,6000,100):
    runtime = runSortReturnTime(j)
    X.append(j)
    Y.append(runtime)

x = np.array(X)
y = np.array(Y)
a, b, c = np.polyfit(x, y, 2)
    
plt.plot(X,Y)
plt.plot(x, a*x**2 +b*x+c)
```
